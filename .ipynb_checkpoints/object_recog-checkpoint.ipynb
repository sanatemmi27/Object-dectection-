```python
import numpy as np
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
import dlib
import face_recognition
import imutils
from imutils import face_utils
from imutils.face_utils import FaceAligner


# Capture video from file
video = cv2.VideoCapture('CCTV.mp4')
video.set(cv2.CAP_PROP_POS_FRAMES, 30) #to start the video file from certain time or skipping frames

count = 0
FPS = 30

while True:

    ret, frame = video.read()
    frame = cv2.resize(frame, (600,600))
    #cv2.line(frame, (300,230), (450,230), (0,255,0), 5)

    if count%FPS == 0:
        print(count%FPS)

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        bbox, label, conf = cv.detect_common_objects(frame)
        print(bbox)
        image = draw_bbox(frame, bbox, label, conf, write_conf= True)

        
        cv2.imshow('frame',frame)
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break

#     else:
#         break
    count+= 1
    

video.release()
cv2.destroyAllWindows()

```
